```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load data
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X_train=std.fit_transform(X_train)
X_test=std.transform(X_test)
```

**Building neural network**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
tf.random.set_seed(3) #the initial weight take the same value
```

```python
#seeting layers
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #input neuronne
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')#output layers
])

```

```python
#compilling neural network
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#training neural network
history=model.fit(X_train,Y_train,validation_split=0.1,epochs=10)
```

```python
y_predict=model.predict(X_test)
```

```python
# Convert probabilities to class predictions
y_pred_classes = np.argmax(y_predict, axis=1)

# Now you can use accuracy_score
accuracy = accuracy_score(Y_test, y_pred_classes)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred_classes)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred_classes))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
```
